Импорт библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import sympy as sp
from scipy.optimize import curve_fit
```

Обработка лабораторной работы

```python
x = 10
n = 21
h_n = np.array([2 / 2 ** i for i in range(1, n + 1)])
```

```python
def diff_1(func, x, h):
    return (func(x + h) - func(x)) / h

def diff_2(func, x, h):
    return (func(x) - func(x - h)) / h

def diff_3(func, x, h):
    return (func(x + h) - func(x - h)) / (2 * h)

def diff_4(func, x, h):
    return 4.0 / 3.0 * (func(x + h) - func(x - h)) / (2 * h) - 1.0 / 3.0 * (func(x + 2 * h) - func(x - 2 * h)) / (4 * h)

def diff_5(func, x, h):
    return 3.0 / 2.0 * (func(x + h) - func(x - h)) / (2 * h) - 3.0 / 5.0 * (func(x + 2 * h) - func(x - 2 * h)) / (4 * h) + 1.0 / 10.0 * (func(x + 3 * h) - func(x - 3 * h)) / (6 * h)
```

Вычисляем производные:

```python
def sp_func_1(x):
    return sp.sin(x ** 2)

def sp_func_2(x):
    return sp.cos(sp.sin(x))

def sp_func_3(x):
    return sp.exp(sp.sin(sp.cos(x)))

def sp_func_4(x):
    return sp.ln(x + 3)

def sp_func_5(x):
    return (x + 3) ** 0.5
```

```python
x_sym = sp.Symbol('x')

def der_func_1():
    return sp.diff(sp_func_1(x_sym), x_sym)

def der_func_2():
    return sp.diff(sp_func_2(x_sym), x_sym)
    
def der_func_3():
    return sp.diff(sp_func_3(x_sym), x_sym)

def der_func_4():
    return sp.diff(sp_func_4(x_sym), x_sym)

def der_func_5():
    return sp.diff(sp_func_5(x_sym), x_sym)
```

```python
print(der_func_1())
print(der_func_2())
print(der_func_3())
print(der_func_4())
print(der_func_5())
```

Функции и их производные

```python
def func_1(x):
    return np.sin(x ** 2)

def func_2(x):
    return np.cos(np.sin(x))

def func_3(x):
    return np.exp(np.sin(np.cos(x)))

def func_4(x):
    return np.log(x + 3)

def func_5(x):
    return np.sqrt(x + 3)
```

```python
def der_func_1(x):
    return 2 * x * np.cos(x ** 2)

def der_func_2(x):
    return - np.sin(np.sin(x)) * np.cos(x)

def der_func_3(x):
    return - np.exp(np.sin(np.cos(x))) * np.sin(x) * np.cos(np.cos(x))

def der_func_4(x):
    return 1 / (x + 3)

def der_func_5(x):
    return 0.5 / np.sqrt(x + 3)
```

```python
ln_h_n = np.log(h_n)
```

```python
mpl.rcParams['font.size'] = 16                   # Управление стилем, в данном случаем - размером шрифта 
plt.figure(figsize = (8,8), facecolor = "white") # Создаем фигуру

# Подписываем оси и график
plt.title(r"График ошибки вычисления производной для функции $y = sin(x^2)$")
plt.ylabel(r"$ln(\Delta)$")
plt.xlabel(r"$ln(h_n), h_n = 2 / 2^n, n = \overline{1, 21}$")

Delta = np.log(np.abs(diff_1(func_1, x, h_n) - der_func_1(x)))
plt.plot(ln_h_n, Delta, '.-', markersize = 9, label = 'Первый метод')

Delta = np.log(np.abs(diff_2(func_1, x, h_n) - der_func_1(x)))
plt.plot(ln_h_n, Delta, '.-', markersize = 9, label = 'Второй метод')

Delta = np.log(np.abs(diff_3(func_1, x, h_n) - der_func_1(x)))
plt.plot(ln_h_n, Delta, '.-', markersize = 9, label = 'Третий метод')

Delta = np.log(np.abs(diff_4(func_1, x, h_n) - der_func_1(x)))
plt.plot(ln_h_n, Delta, '.-', markersize = 9, label = 'Четвертый метод')

Delta = np.log(np.abs(diff_5(func_1, x, h_n) - der_func_1(x)))
plt.plot(ln_h_n, Delta, '.-', markersize = 9, label = 'Пятый метод')

plt.grid(visible = True, which = 'major', axis = 'both', alpha = 1, linewidth = 0.9)   # Активируем сетку
plt.grid(visible = True, which = 'minor', axis = 'both', alpha = 0.5, linestyle = ':')

plt.minorticks_on()
plt.tight_layout()
plt.legend(loc='best', fontsize = 12) # Активируем легенду графика
plt.show()
```
